```python
# load the dependencies
import pandas as pd
```

```python
# Read the csv input file

data = '../Resources/loans_data.csv'

loans_df = pd.read_csv(data)
loans_df.head(15)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
loans_df['month'] = le.fit_transform(loans_df['month'])
loans_df['education'] = le.fit_transform(loans_df['education'])
loans_df['gender'] = le.fit_transform(loans_df['gender'])
```
